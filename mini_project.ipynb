```python
# Unique values in categorical columns
print("\n--- Unique Values ---")
print("States:", df['State'].nunique())
print("Crime Types:", df['Crime_Type'].unique())
print("Rural Area (Yes/No):", df['Rural_Area'].unique())
```

```python
# Total crimes per state
state_crime_counts = df.groupby('State')['Number_of_Incidents'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
state_crime_counts.plot(kind='bar')
plt.title('Total Number of Incidents per State')
plt.xlabel('State')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Crime trends over the years
yearly_trend = df.groupby('Year')['Number_of_Incidents'].sum()
plt.figure(figsize=(10, 5))
yearly_trend.plot(marker='o')
plt.title('Crime Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.grid(True)
plt.show()
```

```python
# Crime type distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Crime_Type', order=df['Crime_Type'].value_counts().index)
plt.title('Crime Type Distribution')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Conviction rate analysis
conviction_rate = df['Convictions'].sum() / df['Number_of_Incidents'].sum()
print(f"\nOverall Conviction Rate: {conviction_rate:.2%}")
```

```python
# Gender-based victim analysis
gender_victims = df[['Victims_Male', 'Victims_Female']].sum()
gender_victims.plot(kind='bar', color=['blue', 'pink'])
plt.title('Total Victims by Gender')
plt.ylabel('Number of Victims')
plt.show()
```

```python
# Rural vs Urban crime
rural_counts = df['Rural_Area'].value_counts()
sns.barplot(x=rural_counts.index, y=rural_counts.values)
plt.title('Crime Occurrences: Rural vs Urban')
plt.ylabel('Number of Records')
plt.show()
```
